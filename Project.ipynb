```python
#CREATE A COLUMN 'SALES'
```

```python
df['SALES'] = df.apply(lambda row : row['QUANTITYORDERED'] * row['PRICEEACH'],axis=1)
df['SALES'] = df['SALES'].round(2)
```

```python
#REARRANGE THE COLUMNS
```

```python
column_names_reordered = ['ORDERNUMBER','QUANTITYORDERED','PRICEEACH','SALES','ORDERLINENUMBER',
                          'ORDERDATE','STATUS','YEAR_ID',	'PRODUCTLINE','PRODUCTCODE','CUSTOMERNAME',
                          'CITY','STATE','COUNTRY','DEALSIZE']
```

```python
df = df[column_names_reordered]
```

```python
df.head()
```

```python
#FIND YEAR WITH HIGHEST SALES
```

```python
df1 = df.groupby('YEAR_ID')['SALES'].sum().reset_index()
df1
```

```python
sns.barplot(data=df1, y='SALES', x='YEAR_ID',hue='YEAR_ID')
plt.title('Sales vs Year')
plt.show()
```

```python
#CREATE COLUMN 'MONTH_ID' HAVING MONTH VALUE BY EXTRACTING MONTH VALUE FROM 'ORDERTIME' COLUMN
```

```python
type(df['ORDERDATE'][0])
```

```python
df['ORDERDATE'] = pd.to_datetime(df['ORDERDATE'])
```

```python
def month_value(date):
    return date.month
```

```python
df['MONTH_ID'] = df['ORDERDATE'].apply(month_value)
```

```python
df.head()
```

```python
#FIND THE MONTH WITH MOST SALES
```

```python
df2 = df.groupby('MONTH_ID')['SALES'].sum().reset_index()
df2.head()
```

```python
sns.barplot(data=df2, y='SALES', x='MONTH_ID')
plt.title('Sales vs Month')
plt.show()
```

```python
#FIND COUNTRY WITH HIGHEST SALES IN EACH YEAR
```

```python
unique_countries = df['COUNTRY'].unique()
temp=pd.DataFrame()
for i, country in enumerate(unique_countries):
    country_data = df[df['COUNTRY'] == country]
    country_data=country_data.groupby('YEAR_ID')['SALES'].sum().reset_index()
    country_data['COUNTRY'] = country
    temp = pd.concat([temp,country_data], ignore_index=True)
temp.head()
```

```python
def calc_max(group):
    df8 = group[group['SALES'] == group['SALES'].max()]
    return df8[['SALES','COUNTRY']]
```

```python
res = temp.groupby('YEAR_ID').apply(calc_max)
res.reset_index(level='YEAR_ID', inplace = True)
res
```

```python
#DISPLAY COUNTRY REVENUE BY YEAR
```

```python
unique_countries = df['COUNTRY'].unique()
temp=pd.DataFrame()
for i, country in enumerate(unique_countries):
    country_data = df[df['COUNTRY'] == country]
    country_data=country_data.groupby('YEAR_ID')['SALES'].sum().reset_index()
    country_data['COUNTRY'] = country
    temp = pd.concat([temp,country_data], ignore_index=True)
plt.figure(figsize=(20,8))
plt.title("Country Revenue by Year")
sns.barplot(x='COUNTRY',y='SALES',data=temp,hue='YEAR_ID')
plt.show()
```

```python
#FIND COUNTRY WITH HIGHEST SALES GROWTH IN 2004
```

```python
unique_countries = df['COUNTRY'].unique()
temp=pd.DataFrame()
for i, country in enumerate(unique_countries):
    country_data = df[df['COUNTRY'] == country]
    country_data=country_data.groupby('YEAR_ID')['SALES'].sum().reset_index()
    country_data['COUNTRY'] = country
    temp = pd.concat([temp,country_data], ignore_index=True)
temp.head()
```

```python
temp = temp[temp['YEAR_ID'] <= 2004]
temp.head()
```

```python
temp['Growth_Rate'] = temp['SALES'].pct_change(periods=1) * 100
temp.head()
```

```python
temp = temp[temp['YEAR_ID'] == 2004]
temp.head()
```

```python
plt.figure(figsize=(20,6))
sns.lineplot(data=temp, x="COUNTRY", y="Growth_Rate")
plt.title('Growth Rate for every country')
plt.xticks(rotation=45)
plt.show()
```

```python
#FIND TOP 5 CUSTOMERS WITH HIGHEST REVENUE CONTRIBUTION
```

```python
unique_years = df['YEAR_ID'].unique()
for i, year in enumerate(unique_years):
    year_data = df[df['YEAR_ID'] == year]
    year_data=year_data.groupby('CUSTOMERNAME')['SALES'].sum().reset_index()
    year_data['YEAR_ID'] = year
    plt.figure(figsize=(10,3))
    plt.title(str(year)+" Revenue by Customer (TOP 5)")
    year_data =year_data.sort_values(by = "SALES", ascending=False)
    year_data=year_data.head(5)
    sns.barplot(x='CUSTOMERNAME',y='SALES',data=year_data,hue='CUSTOMERNAME')
    plt.xticks(rotation = 45)
    plt.show()
```
